 Generate a 5×4 array with random values between 0 and 1 (float).

● Find the row with the maximum sum of values.

● Normalize each row using the formula: ----------------------------------------------------------------------------------------

```python
B=np.random.rand(5,4)
B
```

```python
row_sum = np.sum(B,axis=1)
max_row_index = np.argmax(row_sum)
max_row = B[max_row_index]
print(max_row)
```

```python
x = B
min_value = np.min(x,axis=1,keepdims=True)
max_value = np.max(x,axis=1,keepdims=True)
x_normalized = (x-min_value)/(max_value-min_value)
x_normalized
```

5. arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])

● Find all unique elements

● Find the duplicates (elements that appear more than once) -----------------------------------------------------------------------------------------------

```python
arr = np.array([1,3,5,3,7,9,1,5,5])
unique,counts = np.unique(arr,return_counts=True)
duplicates = unique[counts>1]
print("Unique Elements:",unique)
print("Duplicates:",duplicates)
```

● Reshape the result into a 3×3 matrix

● Extract the diagonal elements

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.array([7,8,9])
join = np.concatenate((a,b,c))
print("Joined_array:",join)
```
